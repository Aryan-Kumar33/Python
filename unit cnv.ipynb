```python
def convert_length(value, unit_from, unit_to):
    factors = {
        "m": 1,
        "cm": 0.01,
        "km": 1000
    }
    return value * factors[unit_from] / factors[unit_to]


def convert_weight(value, unit_from, unit_to):
    factors = {
        "kg": 1,
        "g": 0.001,
        "mg": 0.000001
    }
    return value * factors[unit_from] / factors[unit_to]


def convert_temperature(value, unit_from, unit_to):
    if unit_from == "C":
        if unit_to == "F":
            return value * 9/5 + 32
        if unit_to == "K":
            return value + 273.15
    if unit_from == "F":
        if unit_to == "C":
            return (value - 32) * 5/9
        if unit_to == "K":
            return (value - 32) * 5/9 + 273.15
    if unit_from == "K":
        if unit_to == "C":
            return value - 273.15
        if unit_to == "F":
            return (value - 273.15) * 9/5 + 32
```

```python
value = 5
unit_from = "km"
unit_to = "m"

result = convert_length(value, unit_from, unit_to)
result
```

```python
result = convert_weight(5000, "g", "kg")
result
```

```python
result = convert_temperature(100, "C", "F")
result
```
